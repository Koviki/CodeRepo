<b>Handling Null</b>

```python
# importing pandas package
import pandas as pd

# making data frame from csv file
df = pd.read_csv("D:\Data Science\Data Sets\Bad Record Set.csv")
display(df)

# creating bool series True for NaN values
bool_series = pd.isnull(df["Expense"])

# filtering data
# displaying data only with Gender = NaN
print(bool_series)

df[bool_series]
```

```python
bool_series=df['Expense'].notnull()
df[bool_series]
```

```python
#fillna

display(df['Expense'].fillna(10))
df2=df['Expense'].fillna(10)
df['Filled_Exp']=df2
display(df)
```

```python
df1=df.drop(['Filled_Exp'],axis=1)
display(df1)
```

```python
#Filling null values with the previous ones 
#df.fillna(method ='bfill')
df2=df['Expense'].fillna(method ='pad')
df['Filled_Exp']=df2
display(df)
```

```python
df=df.drop(['Filled_Exp'],axis=1)
display(df)
```

```python
#Using Replace
import numpy as np

df['filled_exp']=df['Expense'].replace(to_replace = np.nan, value = -99)
display(df)
```

```python
df=df.drop(['filled_exp'],axis=1)
display(df)
```

```python
import pandas as pd
df=pd.read_csv("D:\Data Science\Data Sets\Sales Data.csv")
display(df)
```

```python
X=df.iloc[:,[1]].values
y=df.iloc[:,2].values
display(X)
display(y)
```

```python
import numpy as np
import matplotlib.pyplot as p
import sklearn.preprocessing

p.scatter(X,y,color='red')
p.title('Annual Fee vs Turn Over')
p.xlabel('Annual Franchise Fee')
p.ylabel('Turn Over')
p.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=2/5, random_state=0)

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
print("Intercept (b0) = %.2f" %regressor.intercept_)
print("b1= %.2f" %regressor.coef_)

y_prediction = regressor.predict(X_test)
```

```python
y_predict = regressor.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
print("Mean Squared Error: %.2f" %mean_squared_error(y_test,y_predict))
print("Root Mean Squared Error: %.2f" %np.sqrt(mean_squared_error(y_test,y_predict)))
```

```python
p.scatter(X,y,color='red')
p.plot(X_train,regressor.predict(X_train), color='green')
p.title('Annual Fee vs Turn Over')
p.xlabel('Annual Franchise Fee')
p.ylabel('Turn Over')
p.show()
```
