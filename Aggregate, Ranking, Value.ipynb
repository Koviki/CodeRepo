```python
#group by using by clause

import pandas as pd

df=pd.read_csv("D:\Data Science\Data Sets\Window_func.csv")
display(df)
```

```python
df.groupby(by='First_nm').sum()
```

```python
#not using any agg function. This can do group by on multiple columns also
df.groupby("First_nm")["Cost"].count()
```

```python
#categorising based on subdivition created randomly
df
```

```python
df['Order_level']=pd.cut(df['Cost'],bins=3,labels=('lower','middle','high'))
display(df)
```

```python
#using diff aggregate function on different columns all at the same time
df.groupby('First_nm').agg({'id_x':min,
                           'Cost':max})
```

```python
#rolling mean,max,sum. In this case the mean was calculated for the last 5 values
#https://www.stratascratch.com/blog/python-window-functions/

df['Rolling_Mean'] = df['Cost'].rolling(5).mean()
display(df)
```

```python
#for the first five values the data were not available so the Nan output was there. But min_period=1 means it will
#start calculating the values even if 5 values were not there initially

df['Rolling_Mean'] = df['Cost'].rolling(5,min_periods=1).mean()
display(df)
```

```python
#Expanding Aggregate.

df['Expanding_mean'] = df['Cost'].expanding(5).sum()
display(df)
```

```python
#cumcount() will row numbers for each groups. the rownumbers will restart once the group ends. The sort_values will sort
#based on that value

df['Row_count'] = df.groupby(['First_nm']).cumcount()+1
df1=df.sort_values(by=['First_nm','Row_count'])
display(df1)
```

```python
#to find the 2nd highest use iloc to sleect the row_count is 2. This is similar to dense rank

df1.loc[df1['Row_count'] == 2]
```

```python
#Rank

import pandas as pd

df=pd.read_csv("D:\Data Science\Data Sets\Leg_ranks.csv")
display(df)
```

```python
#rank and Dense Rank

df['default_rank'] = df['Legs'].rank() #cat and dog are 3.5 because in normal ranking it shlould be 3, 4. but since the value is same so it makes sense to give then the same rank
df['min_rank'] = df['Legs'].rank(method='min') #here cat and dog are given same rank but the next rank 4 is skipped. so spider got 5
df['NA_bottom'] = df['Legs'].rank(method='min', na_option='bottom') #NA will also be considered for ranking but it will be given the last rank
df['descending'] = df['Legs'].rank(method='min', ascending = False) #ranking is other way round

display(df)
```

```python
#Dense Rank. The skipping ranks won't be there. Other options as mentioned above ar also valid with dense rank


df=pd.read_csv("D:\Data Science\Data Sets\Leg_ranks.csv")


df['dense_rank'] = df['Legs'].rank(method='dense')
display(df)
```
